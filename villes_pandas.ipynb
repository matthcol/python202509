module re:
- match, fullmatch
- search ; équivalent pandas .str.contains()
- sub (chercher/remplacer) ; équivalent pandas .str.replace()

```python
regexp = r'sainte?[ -]'
re.match(regexp, 'Saint Etienne') # res: None
```

```python
re.match(regexp, 'Saint Etienne', flags=re.I)
```

```python
villes.loc[
    villes.gentile.isna(),
    ['nom_standard', 'gentile']
]
```

```python
gentile = villes.loc[
    ~villes.gentile.isna()
    & villes.gentile.str.contains(r"[^a-zéèêùàüÿëïôûâîòçñœ0-9, ?<>;:()·•\.\[\]'/&«»\"’-]", case=False),
    # ['nom_standard', 'gentile']
    'gentile'
].iloc[0]
gentile
```

```python
re.sub(r"[^a-zéèêùàüÿëïôûâîòçñœ0-9, ?<>;:()·•\.\[\]'/&«»\"’-]*", "", gentile)
```

```python
re.sub(r"\[.*\]", "", gentile)
```

```python
villes["gentile2"] = villes.gentile.str.replace(r"\[.*\]", "", regex=True)
villes
```
